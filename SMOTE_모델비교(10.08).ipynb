```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('data/except_personal_rehailitation.csv')
data.head()
```

```python
data=data.drop(['Unnamed: 0'],axis=1)
```

### data1 - 필요없는 column 제거된 데이터

```python
# 필요없는 column 제거
data1=data.drop(['application_id', 'user_id','loanapply_insert_time', 'personal_rehabilitation_yn', 'personal_rehabilitation_complete_yn','insert_time','company_enter_month'],axis=1)
```

```python
data1.head()
```

### data2 - data1을 가지고 범주형 더미화

```python
data2 = pd.get_dummies(data1, columns = ['income_type','employment_type','houseown_type','purpose'])
```

```python
data2.head()
```

### train : is_applied null 아닌 값 / test : is_applied null 값(제출 데이터)

```python
ii=data2[data2.is_applied.isnull()].index
train=data2.drop(ii) # 3,4,5월 
test=data2.loc[ii] # 6월(실제 예측해야할 값)
```

```python
# 3,4,5월 데이터의 is_applied 비율
train['is_applied'].value_counts()
```

```python
X = train.drop('is_applied',axis=1)
y = train['is_applied']
```

#### X, y에 대하여 oversampling 수행

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_over, y_over = smote.fit_resample(X,y)

print('SMOTE 적용 전 학습용 피처/레이블 데이터 세트: ', X.shape, y.shape)
print('SMOTE 적용 후 학습용 피처/레이블 데이터 세트: ', X_over.shape, X_over.shape)
print('SMOTE 적용 후 레이블 값 분포: \n', pd.Series(y_over).value_counts())
```

```python
# data_over은 X,y oversampling된 거 합쳐진 data
data_over = pd.concat([X_over,y_over],axis = 1)
```

```python
data_over.head()
```

### X_over, y_over 데이터셋 분리 -> train_X, train_y로 훈련, test_X, test_y로 평가

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X_over, y_over, test_size = 0.3, random_state = 42 )
```

```python
train_y.value_counts()
```

### 모델 평가 성능지표

```python
from sklearn import metrics
from sklearn.metrics import r2_score, mean_squared_error, accuracy_score, f1_score, precision_score, recall_score, roc_auc_score
```

```python
def metrics(test_y, pred):
    accuracy = accuracy_score(test_y, pred)
    precision = precision_score(test_y, pred)
    recall = recall_score(test_y, pred)
    f1 = f1_score(test_y, pred)
    roc_score = roc_auc_score(test_y, pred, average='macro')
    print('정확도 : {0:.2f}, 정밀도 : {1:.2f}, 재현율 : {2:.2f}'.format(accuracy, precision, recall))
    print('f1-score : {0:.2f}, auc : {1:.2f}'.format(f1, roc_score, recall))
```

### LogisticRegression

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(train_X, train_y)
pred = lr.predict(test_X)

metrics(test_y, pred)
```

### RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(train_X, train_y)
pred = rf.predict(test_X)

metrics(test_y, pred)
```

### DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(train_X, train_y)
pred = dt.predict(test_X)

metrics(test_y, pred)
```

### ExtraTreesClassifier

```python
# 이건 아직 예전꺼로 돌린거라 성능 낮게 써있어요
from sklearn.ensemble import ExtraTreesClassifier

et = ExtraTreesClassifier()
et.fit(train_X, train_y)
pred = et.predict(test_X)

metrics(test_y, pred)
```

### XGBoost

```python
from xgboost import XGBClassifier

xgb = XGBClassifier()
xgb.fit(train_X, train_y)
pred = xgb.predict(test_X)

metrics(test_y, pred)
```

### CatBoost

```python
# 이건 아직 예전꺼로만 돌려보고 안 돌려봄
from catboost import CatBoostClassifier

cb = CatBoostClassifier()
cb.fit(train_X, train_y)
pred = cb.predict(test_X)

metrics(test_y, pred)
```

### LightGBM

```python
from lightgbm import LGBMClassifier

lgbm = LGBMClassifier()
lgbm.fit(train_X, train_y)
pred = lgbm.predict(test_X)

metrics(test_y, pred)
```
